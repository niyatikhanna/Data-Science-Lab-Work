# 🎓 Student Performance Visualization (Enhanced)

This project provides improved visualizations and insights into student performance.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Dataset
data = {
    "Name": ["A", "B", "C", "D", "E"],
    "Math": [85, 78, 92, 70, 88],
    "Science": [80, 75, 89, 72, 90],
    "English": [78, 82, 85, 68, 87]
}

df = pd.DataFrame(data)
df
```

```python
# Add Average column
df["Average"] = df[["Math", "Science", "English"]].mean(axis=1)
df
```

## 📊 Bar Chart (Subject-wise Performance)

```python
df.set_index("Name")[["Math", "Science", "English"]].plot(kind="bar")
plt.title("Subject-wise Performance")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## 📈 Line Chart (Trend of Marks)

```python
df.set_index("Name")[["Math", "Science", "English"]].plot(kind="line", marker='o')
plt.title("Marks Trend")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.grid()
plt.show()
```

## 🥧 Pie Chart (Average Distribution)

```python
plt.pie(df["Average"], labels=df["Name"], autopct='%1.1f%%')
plt.title("Average Marks Distribution")
plt.show()
```

## 🔍 Insights

```python
top_student = df.loc[df['Average'].idxmax()]
print("Top Performer:")
print(top_student)
```
